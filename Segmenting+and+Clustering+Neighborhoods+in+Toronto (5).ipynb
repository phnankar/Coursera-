# Segmenting and Clustering Neighborhoods in Toronto

```python
import pandas as pd
from bs4 import BeautifulSoup 
import requests
!pip install folium
import folium
```

 <b>Scrapping of table from wikipedia page

```python
source = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
```

```python
soup = BeautifulSoup(source,"xml")
```

```python
table  = soup.find('table')
```

<b> Create dataframe with three columns: PostalCode, Borough, and Neighborhood

```python
df = pd.DataFrame(columns = ['PostalCode', 'Borough', 'Neighborhood'])
```

```python
for tr in table.find_all('tr'):
    row = []
    for td in tr.find_all('td'):
        row.append(td.text.strip())
    if(len(row)):
        df.loc[len(df)] = row
```

```python
df
```

<b> Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
df = df[df['Borough'] != 'Not assigned']
df
```

<b> Neighborhoods separated with a comma 

```python
df["Neighborhood"]= df["Neighborhood"].str.replace("/", ",", case = False) 
df
```

<b> No of rows in the dataframe

```python
df.shape
```

<b> Read geo data from csv

```python
df_geo = pd.read_csv("http://cocl.us/Geospatial_data")
```

```python
df_geo.rename(columns={'Postal Code':'PostalCode'},inplace=True)
```

```python
df_geo
```

<b> Merge both dataframes

```python
data = pd.merge(df,df_geo,on = "PostalCode")
```

```python
data
```

<b> Dataframe containing neighborhoods in Toronto  

```python
tmp = data[data['Borough'].str.contains("Toronto")]
```

```python
tmp
```

```python
X = tmp[["Latitude","Longitude"]]
```

<b> KMeans Clusteing using location

```python
from sklearn.cluster import KMeans
k = 5
kmeans = KMeans(n_clusters=k, random_state=0).fit(X)

kmeans.labels_
```

```python
tmp['Cluster'] = kmeans.labels_
```

```python
tmp.sort_values("Cluster")
```

<b> Display of neighbourhood clusters of Torronto on Map using colors

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
map_clusters = folium.Map(location=[43.648198,-79.379817], zoom_start=11)
# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(tmp['Latitude'], tmp['Longitude'], tmp['Neighborhood'], tmp['Cluster']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
